```python
df = df.rename(columns = {'Total Charge' : 'Total Charges'})
```

```python
df['Markup'] = df['Total Charges'] / df['Total Costs']
```

```python
net = df[['Facility Name', 'Total Charges', 'Total Costs']]
net.head(2)
```

```python
netgroup = net.groupby('Facility Name').aggregate(['sum'])
netgroup.head(2)
```

```python
netgroup = net.groupby('Facility Name').sum()
netgroup.head(2)
```

```python
netgroup['Net Income'] = netgroup['Total Charges'] - netgroup['Total Costs']
```
